## Retrieving Products on Amazon

```python
import pandas as pd
import requests
import json

# set up the request parameters
params = {
  'api_key': '1011DCD5DDD145908232408370D246DD',
  'type': 'product',
  'amazon_domain': 'amazon.com',
  'asin': 'B06XWZWYVP'
}

# make the http GET request to Rainforest API
api_result_product = requests.get('https://api.rainforestapi.com/request', params)

#print the JSON response from Rainforest API
print(json.dumps(api_result_product.json()))
```

```python
#check the keys
api_result_product.json().keys()
```

```python
#populate the data into a dataframe
df_also_viewed = pd.DataFrame(api_result_product.json()['also_viewed'])
df_also_viewed
```

```python
#populate the data into a data frame
df_sponsored_products = pd.DataFrame(api_result_product.json()['sponsored_products'])
df_sponsored_products
```
